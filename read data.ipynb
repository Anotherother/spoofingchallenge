# read data ASVSpoof 2017 database



Welcome to participate the automatic speaker verification spoofing and countermeasures challenge (ASVspoof2017). 

Here is a description of the development data (including training data).

1. ./wav directory contains audio files used for training and development. 
    The waveforms in the ./wav directory are in RIFF/WAVE format. The sampling rate is 16 kHz, and stored in 16-bit format.

    
2. ./protocol directory contains protocol files in ASCII format for training and development.
    ./protocol/ASVspoof2017_train.trn   : contains file list to be used to train human and spoofed speech detectors.
    ./protocol/ASVspoof2017_dev.trl : development trial list to validate the spoofing detector.

    The two file have the same format and the meaning of each column is listed as follows:
        1st column: unique file ID
        2nd column: speech type identifier: genuine means the trial is original speech; spoof means the file is created with replay attack.
        3rd column: speaker ID
        4th column: RedDots common phrase ID
        5th column: Environment ID ('-' for genuine speech)
        6th column: Playback device ID ('-' for genuine speech)
        7th column: Recording device ID ('-' for genuine speech)

    The file for each training sample can be located at ./wav/train/unique_file_ID.wav
    The file for each development trial sample can be located at ./wav/dev/unique_file_ID.wav

```python
import pandas as pd

dev_file = "./protocol/ASVspoof2017_dev.trl"
train_file = './protocol/ASVspoof2017_train.trn'

# train set   
with open (train_file, "r") as myfile:
    data_train=myfile.read().split('\n')
    
for x in  range(len(data_train)): 
    data_train[x] =  data_train[x].split(' ')

# dev set
with open (dev_file, "r") as myfile:
    data_dev=myfile.read().split('\n')
     
for x in  range(len(data_dev)): 
    data_dev[x] =  data_dev[x].split(' ')
    
columns = ['FileID', 'Type','SpeakerID', 'CommonPhrase',
           'EnvID', 'PlaybackDeviceID', 'RecondDeviceID']

dataframe_train = pd.DataFrame(data_train,columns=columns)
dataframe_dev = pd.DataFrame(data_dev,columns=columns)
```

```python
dataframe_dev.to_csv('dev_label.csv', index = False)
dataframe_train.to_csv('train_label.csv', index = False)
```
